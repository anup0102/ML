```python
import pandas as pd
```

```python
data = pd.read_csv("house_pricing.csv")
```

```python
data.head(20)
```

```python
data = pd.get_dummies(data, columns=['SaleType', 'SaleCondition', 'GarageType', 'PavedDrive'], drop_first=True)
```

```python
data = data.select_dtypes(exclude ='object')
data = data.drop(['GarageYrBlt'], axis=1)
```

```python
data.isnull().any()
```

```python
data.fillna(0, inplace=True)
```

```python
data.head()
```

```python
X = data.drop(['SalePrice'], axis=1).values
y = data['SalePrice'].values


```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lnr = LinearRegression()
```

```python
lnr.fit(X_train, y_train)
y_pred = lnr.predict(X_test)
```

```python
lnr.score(X_train, y_train)
```

```python
lnr.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha = 2, normalize = True)
lasso_model.fit(X_train, y_train)
lasso_model.score(X_train, y_train)
```

```python
y_pred = lasso_model.predict(X_test)
lasso_model.score(X_test, y_test)
```
